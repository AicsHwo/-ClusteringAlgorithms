### 1. K Means

````python
import DataPointGeneration as DPG
mog = DPG.MixtureOfGaussian.Random_ND_Clusters(N = 3, D = 2)
datapoints = mog([32, 52, 73])
mog.display()
````

````python
datapoints_shuffled = DPG.shuffle(datapoints)
````

````python
import ClusteringAlgorithms as CAs
kmeans = CAs.KMeans(k = 3)
cluster_labels = kmeans(datapoints_shuffled)
kmeans.display()
````

````python
%matplotlib nbagg
kmeans_animated = CAs.KMeans(k = 3, maxiter=10)
kmeans_animated.in_animation(datapoints_shuffled)
````
